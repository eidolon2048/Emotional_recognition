# 2. Draw a rectangle across the face

download "HaarCascade" face recognition algoritm here:

https://github.com/opencv/opencv/tree/master/data/haarcascades

```python
# open file
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
```

```python
# Ipmort from 3 to 1 color channel
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = faceCascade.detectMultiScale(gray, 1.1, 4)

# Draw a rectangle
for(x, y, w, h) in faces:
    cv2.rectangle(img,(x, y), (x+w, y+h), (0, 255, 0), 2)

# Show    
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# 3. Real time video

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

#connect to webcam
cam = cv2.VideoCapture(1)
#Check if webcam working correctly
if not cam.isOpened():
   cam = cv2.VideoCapture(0)
if not cam.isOpened():
   raise IOError("Cannot open Webcam")

```

```python
while True:
    ret,frame = cam.read()
    result = DeepFace.analyze(frame, actions = ["emotion"])

    # Draw rectangle across the face
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    for (x, y, w, h) in faces:
        cv2.rectangle(img,(x, y), (x+w, y+h), (0, 255, 0), 2)

    # Add lable of emotion
    font = cv2.FONT_HERSHEY_SIMPLEX

    #use putText() method for inserting text on video
    cv2.putText(img, result,
                    (0, 50), font, 1,
                    (0, 0, 200), 3,
                    cv2.LINE_4);
    cv2.imgshow("Demo video", frame)
    if cv2.waitKey(2) & 0xFF == ord("q"):
       break

cam.realease()
cv2.destroyALLWindows
```
